```python
import pandas as pd
from PIL import Image
import io, os

df = pd.read_parquet("/home/nira/Documents/code/aiml/emoji-dataset/data/train-00000-of-00001-38cc4fa96c139e86.parquet")

out_dir = "/home/nira/Documents/code/aiml/VQ_VAE_Extensive_Research/data"

os.makedirs(out_dir, exist_ok=True)

for i, row in df.iterrows():
    img = Image.open(io.BytesIO(row["image"]["bytes"]))
    img.save(f"{out_dir}/emoji_{i:05d}.png")
    print(f"Saved emoji_{i:05d}.png")

print(f"âœ… Saved {len(df)} emoji images to {out_dir}/")
```
